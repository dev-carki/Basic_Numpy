# axis
- ndarray의 shape는 행 열 높이가 아니라 axis0, 1, 2 단위로 부여됨

```python
import numpy as np
```

```python
array1 = np.array([1, 2, 3, 4])
```

```python
array1.shape
# axis 0이 4
```

```python
array2 = np.array([
    [1, 2, 3, 4],
    [5, 6, 7, 8]
])
```

```python
array2.shape # axis0 = 2, axis1 = 4
```

```python
array3 = np.array([
    [
        [1, 2, 3],
        [4, 5, 6]
    ],
    [
        [1, 2, 3],
        [4, 5, 6]
    ],
    [
        [1, 2, 3],
        [4, 5, 6]
    ],
    [
        [1, 2, 3],
        [4, 5, 6]
    ]
])
```

```python
array3.shape # axis0 = 4, axis1 = 2, axis2 = 3
```

## 행 -> 열 -> 높이 순으로 axis마지막, axis마지막-1, axis0
