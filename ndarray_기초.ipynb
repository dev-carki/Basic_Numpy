```python
import numpy as np
```

# 1차원
## .ndim
- 차원 반환

## .shape
- (행, 열) 튜플 형태로 반환

```python
array1 = np.array([1,2,3])
```

```python
array1.ndim
```

```python
array1.shape
```

# 2차원
- (2, 3) -> (행, 열) 튜플 형태로 반환

```python
array2 = np.array([
    [1, 2, 3],
    [3, 4, 5]
])
```

```python
array2.ndim
```

```python
array2.shape
```

# 3차원

```python
array3 = np.array([[
    [1, 2, 3],
    [3, 4, 5]
]])
```

```python
array3.ndim
```

```python
array3.shape # 2행 3열이 1개
```

```python
array4 = np.array([
    [
        [1, 2, 3],
        [4, 5, 6]
    ],
    [
        [2, 4, 6],
        [4, 5, 6]
    ]
])
```

```python
array4.shape # 2행 3열이 2개
```
